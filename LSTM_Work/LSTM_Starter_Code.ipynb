```python
## split entire tensor into prior-X and target

tensor_arr = []
targets = []
for i in range(5):
  tensor_arr.append(torch.tensor(init_10_samples[i:i+5]).float())
  targets.append(torch.tensor(init_10_samples[i+6]).float())

print(tensor_arr)
print(targets)

## split above sets into training and validation and testing?
```

```python
from torch.utils.data import TensorDataset, DataLoader
```

```python
tensor_arr[0].unsqueeze(0).shape
```

```python
targets[0].unsqueeze(0).shape
```

```python
## train_data needs to add all remaining data

train_data = TensorDataset(tensor_arr[0].unsqueeze(0), targets[0].unsqueeze(0))
```

```python
train_loader = DataLoader(train_data, batch_size = 1)
```

```python
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

class LSTM_Model(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(LSTM_Model, self).__init__()
        self.hidden_size = hidden_size
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.linear = nn.Linear(hidden_size, output_size)
    def forward(self, x):
        lstm_out, (h_n, c_n) = self.lstm(x)
        last_hidden_state = h_n[0]
        x = self.linear(last_hidden_state)
        return x
```

```python
model = LSTM_Model(289, 50, 289)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
train_loss = []
valid_loss = []

def train_model(model, train_loader, criterion, optimzer, num_epochs):

  for epoch in range(num_epochs):
      model.train()
      curr_epoch_train_loss = 0
      for sequences, targets in train_loader:
        optimizer.zero_grad()
        outputs = model(sequences)
        loss = criterion(outputs, targets)

        loss.backward()
        optimizer.step()
        curr_epoch_train_loss += loss.item()
      ## do validation loss
      train_loss.append(curr_epoch_train_loss)
      model.eval()
      curr_epoch_valid_loss = 0
      for sequences, targets in valid_loader:
        outputs = model(sequences)
        loss = criterion(outputs, targets)
        curr_epoch_valid_loss += loss.item()
      valid_loss.append(curr_epoch_valid_loss)
      print('Epoch:', epoch, 'training loss:', curr_epoch_train_loss, 'valid loss:', curr_epoch_valid_loss)

train_model(model, train_loader,criterion, optimizer, 1000)
```
