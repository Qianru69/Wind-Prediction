```python
simp_pyrecast_80m_df

from haversine import haversine
import math

def find_closest_point(df, point):
  closest_idx = 0
  closest_long = None
  closest_lat = None
  closest_dist = math.inf
  point_lat = point[1]
  point_long = point[0]
  for idx, row in df.iterrows():
    df_lat = row['latitude']
    df_long = row['longitude']
    if row['longitude'] > 180:
      df_long = row['longitude'] - 360
    curr_dist = haversine((point_lat, point_long), (df_lat, df_long))
    if curr_dist < closest_dist:
      closest_dist = curr_dist
      closest_idx = idx
      closest_long = df_long
      closest_lat = df_lat

  return closest_idx, closest_long, closest_lat

find_closest_point(simp_pyrecast_80m_df, (-123.25, 38.0))
```

```python
import math

# for each 17x17 area, find closest point in simp_pyrecast_80_df

pyrecast_diff_df = pd.DataFrame(columns=own_model_longitude_range, index=reversed(own_model_latitude_range))

closest_points_list = []

for lon in own_model_longitude_range:
  for lat in own_model_latitude_range:
    idx, longitude, latitude = find_closest_point(simp_pyrecast_80m_df, (lon, lat))
    closest_points_list.append((idx, longitude, latitude))
    curr_df = simp_pyrecast_80m_df.loc[idx,:]
    # print('returned long:', longitude, 'returned lat:', latitude, 'curr lon:', lon, 'curr lat', lat, 'curr_df ws:', curr_df['ws'])
    # print(type(curr_df['ws']))
    pyrecast_df_with_info.loc[lat, lon] = curr_df['ws']
    # print(pyrecast_df_with_info)
    pyrecast_diff_df.loc[lat, lon] = curr_df['ws'] - predictions_df.iloc[predictions_df.index.get_loc(lat), predictions_df.columns.get_loc(lon)]

pyrecast_df_with_info
# pyrecast_diff_df
# pyrecast_diff_df.to_csv('/content/drive/My Drive/CS 663 - Project/own_model_vs_pyrecast_df.csv')
```
